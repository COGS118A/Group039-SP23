# Evaluation Metrics

1. **Accuracy** :

Accuracy measures the proportion of correctly classified instances over the total number of instances. In our case, accuracy represents the percentage of correct predictions made by the model.

Example: If the model predicts 700 out of 1000 instances correctly, the accuracy would be 70%

2. **ROC-AUC** :

ROC-AUC evaluates the tradeoff between the true positive rate and the false positive rate by calculating the area under the receiver operating characteristic curve. It provides a measure of how well the model can distinguish between positive and negative instances.

Example: A ROC-AUC of 0.75 means that the model can correctly rank 75% of positive instances higher than negative instances

3. **Precision-recall** :

This curve will show the tradeoff between the y axis precision (TP / (TP + FP)) and the x axis recall (TP / (TP + FN). We want to avoid as many FN as possible so that people do get the treatments they need. We then expect recall to stay high (right side of the x axis) while the precision varies (up and down the y axis) as we change the algorithms and models.

## Imports

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.metrics import classification_report
from sklearn.compose import ColumnTransformer
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Subsection 1 - Data cleaning


### Dataset

```python
dataset = pd.read_csv('diabetes_prediction_dataset.csv')
dataset
```

### EDA by pairplot

```python
sns.pairplot(dataset, hue='diabetes')
```

### One hot encoding on categorical variables

Encode gender and smoking history. These two are categorical variables , therefore we will be using one-hot encoding.

pandas.get_dummies() is used for data manipulation. It converts categorical data into dummy or indicator variables.

```python
#encode categorical variables
dataset_encoded = pd.get_dummies(dataset, columns=['gender', 'smoking_history'])
dataset_encoded
```

Split the dataset into features "X" and the target variable "y"

```python
# input all fetures except diabetes, drop diabetes column:
X = dataset_encoded.drop('diabetes', axis=1)

# y taget value - what we want to predict
y = dataset_encoded['diabetes']
```

X represents the feature data and y represents the corresponding target labels.

Split the data into trainning and testing sets. Because there are 100,000 points we decide to do split the data into 80% training and 20% testing sets.

Because 'age', 'bmi', 'HbA1c_level', 'blood_glucose_level' are continuos variables we normalize them to ensure they have similar ranges.


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

Initialize the standard scaler and scale the features

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Subsection 2 - SVM

Support Vector Machines (SVM) is a supervised learning algorithm used for classification tasks. By utilizing SVM, we aim to maximize the separation between diabetic and non-diabetic individuals, making it a powerful tool for predicting diabetes based on various features provided in the dataset.


Initialize and train the SVM classifier

```python
svm = SVC()
svm.fit(X_train_scaled, y_train)
```

Accuracy: Evaluate the SVM classifier on the test set

```python
accuracy = svm.score(X_test_scaled, y_test)
print("Accuracy:", accuracy)
```

```python
#make predictions on the test set
X_test_scaled = scaler.transform(X_test)
y_pred = svm.predict(X_test_scaled)
y_prob = svm.decision_function(X_test_scaled)
```

```python
svm_report = classification_report(y_test, y_pred)
print("Classification Report using SVM:")
print(svm_report)
```

Calculate ROC-AUC

```python
# ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = roc_auc_score(y_test, y_prob)

##plot curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC curve (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

Calculate Precision-Recall

```python
#calculate precision-recall curve and average precision
precision, recall, _ = precision_recall_curve(y_test, y_prob)
avg_precision = average_precision_score(y_test, y_prob)

#Plot curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Precision-Recall curve (Avg Precision = {:.2f})'.format(avg_precision))
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()
```

## Subsection 5 - Nested Cross-Validation


Because the dataset is very big the nested cross validation is taking so long to run, therefore we decide to randomly select a subset of 25,000 observations.

```python
subset_size = 25000
subset_indices = np.random.choice(dataset.shape[0], subset_size, replace=False)
subset_dataset = dataset.iloc[subset_indices]

# Encode categorical variables
subset_dataset_encoded = pd.get_dummies(subset_dataset, columns=['gender', 'smoking_history'])

#Separate features and target again, using subset now.
X = subset_dataset_encoded.drop('diabetes', axis=1)
y = subset_dataset_encoded['diabetes']
```

For the algorithms in nested cross-validation, we are considering SVM, Logistic Regression, and Decission Trees.

Create a pipeline for each algorithm.


```python
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])

lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression())
])

dt_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dt', DecisionTreeClassifier())
])
```

Define the hypermarameters for each algorithm

```python
svm_param_grid = {
    'svm__C': [0.1, 1, 10],
    'svm__kernel': ['linear', 'rbf']
}

lr_param_grid = {
    'lr__C': [0.1, 1,10],
    'lr__penalty': ['l2']
}

dt_param_grid = {
    'dt__max_depth': [None, 5, 10],
    'dt__min_samples_split': [2, 5, 10]
}
```

Perform nested cross-validation for each algorithm scoring accuracy using GridSearchCV:

We are performing nested cross-validation. The outer loop of cross-validation is performed using cross_val_score, where we pass the model (svm_grid, lr_grid, dt_grid) as the estimator. The inner loop of cross-validation for hyperparameter tuning is handled by the GridSearchCV.

```python
#THIS WILL TAKE TIMEE, creating the subset helped but still takes a little of time.

#Perform nested cross-validation for SVM
svm_grid = GridSearchCV(svm_pipeline, svm_param_grid, cv=3, scoring='accuracy')
svm_scores = cross_val_score(svm_grid, X, y, cv=3, scoring='accuracy')

#Perform nested cross-validation for Logistic Regression
lr_grid = GridSearchCV(lr_pipeline, lr_param_grid, cv=3, scoring='accuracy')
lr_scores = cross_val_score(lr_grid, X, y, cv=3, scoring='accuracy')

#Perform nested cross-validation for Decision Trees
dt_grid = GridSearchCV(dt_pipeline, dt_param_grid, cv=3, scoring='accuracy')
dt_scores = cross_val_score(dt_grid, X, y, cv=3, scoring='accuracy')
```

Output best parameters for SVM

```python
svm_grid.fit(X, y)
print(svm_grid.best_params_)
```

Output best parameters for Logistic Regression

```python
lr_grid.fit(X, y)
print(lr_grid.best_params_)
```

Output best parameters for Decision Trees

```python
dt_grid.fit(X, y)
print(dt_grid.best_params_)
```

From the nested cross-validation process print the mean accuracy scores for each algorithm

```python
# Print the mean accuracy scores for each algorithm
print("Support Vector Machine (SVM):")
print("Mean Score:", svm_scores.mean())
print("Scores:", svm_scores)
print()

print("Logistic Regression (LR):")
print("Mean Score:", lr_scores.mean())
print("Scores:", lr_scores)
print()

print("Decision Tree (DT):")
print("Mean Score:", dt_scores.mean())
print("Scores:", dt_scores)
print()
```

### Nested Cross-Validation by Recall Metric

```python
#DOING IT BY RECALL METRIC: 
#Perform nested cross-validation for SVM
svm_grid = GridSearchCV(svm_pipeline, svm_param_grid, cv=3, scoring='recall')
svm_scores = cross_val_score(svm_grid, X, y, cv=3, scoring='recall')

#Perform nested cross-validation for Logistic Regression
lr_grid = GridSearchCV(lr_pipeline, lr_param_grid, cv=3, scoring='recall')
lr_scores = cross_val_score(lr_grid, X, y, cv=3, scoring='recall')

#Perform nested cross-validation for Decision Trees
dt_grid = GridSearchCV(dt_pipeline, dt_param_grid, cv=3, scoring='recall')
dt_scores = cross_val_score(dt_grid, X, y, cv=3, scoring='recall')



# Print the mean accuracy scores for each algorithm
print("Support Vector Machine (SVM):")
print("Mean Recall Score:", svm_scores.mean())
print("Scores:", svm_scores)
print()

print("Logistic Regression (LR):")
print("Mean Recall Score:", lr_scores.mean())
print("Scores:", lr_scores)
print()

print("Decision Tree (DT):")
print("Mean Recall Score:", dt_scores.mean())
print("Scores:", dt_scores)
print()
```
